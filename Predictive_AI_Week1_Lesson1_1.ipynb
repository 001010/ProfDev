```python
# Traditional Programming Example
def classify_temperature(temp):
    """Old-school programming with explicit rules"""
    if temp > 80:
        return "It's hot! üî•"
    elif temp > 60:
        return "It's warm ‚òÄÔ∏è"
    else:
        return "It's cold ‚ùÑÔ∏è"

# Test it
temperatures = [45, 72, 85, 55, 90]
print("Traditional Programming Approach:")
for temp in temperatures:
    print(f"{temp}¬∞F - {classify_temperature(temp)}")
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

# Machine Learning Example
# Training data: [temperature] and labels
X_train = np.array([[45], [52], [58], [68], [75], [82], [88], [95]])
y_train = np.array(['cold', 'cold', 'warm', 'warm', 'warm', 'hot', 'hot', 'hot'])

# Create and train the model (it learns the pattern!)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Now predict new temperatures
test_temps = [[50], [72], [85]]
predictions = model.predict(test_temps)

print("\nMachine Learning Approach:")
for temp, prediction in zip(test_temps, predictions):
    emoji = '‚ùÑÔ∏è' if prediction == 'cold' else '‚òÄÔ∏è' if prediction == 'warm' else 'üî•'
    print(f"{temp[0]}¬∞F - Predicted: {prediction} {emoji}")

print("\nNotice: We never wrote rules! The model learned them from examples.")
```

```python
# Import the libraries we need (already installed in Colab!)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

# Verify everything loaded
print("‚úì All libraries loaded successfully!")
print("‚úì Ready to build your first model!")
```

```python
# Our training data: hours studied and the resulting test scores
hours_studied = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
test_scores = np.array([50, 55, 65, 70, 75, 82, 88, 95])

print("üìä Training Data:")
print("=" * 25)
print("Hours | Score")
print("-" * 25)
for hours, score in zip(hours_studied, test_scores):
    print(f"  {hours[0]}   |  {score}")
print("=" * 25)
```

```python
# Create a professional visualization
plt.figure(figsize=(10, 6))
plt.scatter(hours_studied, test_scores, color='blue', s=150, alpha=0.6,
            edgecolors='darkblue', linewidths=2, label='Actual Data')
plt.xlabel('Hours Studied', fontsize=13, fontweight='bold')
plt.ylabel('Test Score', fontsize=13, fontweight='bold')
plt.title('Study Hours vs Test Scores', fontsize=15, fontweight='bold', pad=15)
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=11)
plt.xlim(0, 9)
plt.ylim(45, 100)

# Add some style
plt.tight_layout()
plt.show()

print("\nüîç What do you notice?")
print("  ‚Ä¢ More hours ‚Üí higher scores (positive relationship)")
print("  ‚Ä¢ Points roughly form a straight line (linear relationship)")
print("  ‚Ä¢ No obvious outliers or weird patterns")
print("  ‚úì This data is perfect for Linear Regression!")
```

```python
# Create a Linear Regression model
model = LinearRegression()

print("ü§ñ Training the model...")
print("   (The model is learning the pattern...)")

# Train it on our data (this is where it "learns")
model.fit(hours_studied, test_scores)

print("‚úÖ Model trained successfully!")
print("\nüìê The Model's Equation:")
print(f"   Score = {model.intercept_:.2f} + ({model.coef_[0]:.2f} √ó Hours)")
print(f"\nüß† What this means:")
print(f"   ‚Ä¢ Base score (0 hours): {model.intercept_:.1f} points")
print(f"   ‚Ä¢ Each additional hour adds: {model.coef_[0]:.1f} points")
```

```python
# Predict scores for different study hours
new_hours = np.array([[2.5], [5.5], [9]])  # 2.5, 5.5, and 9 hours

print("üîÆ Making predictions for new students...")
predictions = model.predict(new_hours)

print("\nüìä Predictions:")
print("=" * 35)
print("Study Hours | Predicted Score")
print("-" * 35)
for hours, score in zip(new_hours, predictions):
    print(f"   {hours[0]:4.1f}     |     {score:5.1f}")
print("=" * 35)

# Interesting insights
print(f"\nüí° Insights:")
print(f"  ‚Ä¢ Student studying 2.5 hrs ‚Üí expects ~{predictions[0]:.0f} points")
print(f"  ‚Ä¢ Student studying 5.5 hrs ‚Üí expects ~{predictions[1]:.0f} points")
print(f"  ‚Ä¢ Student studying 9 hrs ‚Üí expects ~{predictions[2]:.0f} points")
```

```python
# Create a comprehensive visualization
plt.figure(figsize=(12, 6))

# Plot actual training data
plt.scatter(hours_studied, test_scores, color='blue', s=150,
           alpha=0.7, edgecolors='darkblue', linewidths=2,
           label='Actual Training Data', zorder=3)

# Plot the prediction line
hours_range = np.linspace(0, 10, 100).reshape(-1, 1)
predicted_scores = model.predict(hours_range)
plt.plot(hours_range, predicted_scores, color='red', linewidth=3,
        label='Model Prediction Line', zorder=2)

# Plot new predictions
plt.scatter(new_hours, predictions, color='gold', s=300,
           marker='*', edgecolors='orange', linewidths=2,
           label='New Predictions', zorder=4)

# Styling
plt.xlabel('Hours Studied', fontsize=13, fontweight='bold')
plt.ylabel('Test Score', fontsize=13, fontweight='bold')
plt.title('Study Hours vs Test Scores - Model Performance',
          fontsize=15, fontweight='bold', pad=15)
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=11, loc='lower right')
plt.xlim(0, 10)
plt.ylim(40, 105)

plt.tight_layout()
plt.show()

print("* Gold stars show predictions for new students!")
```

```python
# Make predictions on our training data
train_predictions = model.predict(hours_studied)

# Calculate metrics
r2 = r2_score(test_scores, train_predictions)
mae = mean_absolute_error(test_scores, train_predictions)

print("üìà Model Performance Metrics:")
print("=" * 50)
print(f"R¬≤ Score: {r2:.3f}")
print(f"  ‚Üí Interpretation: {r2*100:.1f}% of score variation explained")
print(f"  ‚Üí Rating: {'Excellent' if r2 > 0.9 else 'Good' if r2 > 0.7 else 'Needs Work'}")
print()
print(f"Mean Absolute Error (MAE): {mae:.2f} points")
print(f"  ‚Üí Interpretation: Predictions off by ~{mae:.1f} points on average")
print(f"  ‚Üí That's {(mae/100)*100:.1f}% error on a 100-point scale")
print("=" * 50)

# Visual comparison
print("\nüéØ Prediction Accuracy Check:")
print("=" * 45)
print("Hours | Actual | Predicted | Difference")
print("-" * 45)
for h, actual, pred in zip(hours_studied, test_scores, train_predictions):
    diff = actual - pred
    symbol = "‚úì" if abs(diff) < 3 else "~"
    print(f"  {h[0]}   |   {actual:2.0f}   |   {pred:5.1f}   | {diff:+5.1f} {symbol}")
print("=" * 45)
```

```python
#@title üè¢ **Business Exercise: Advertising ROI Prediction** { display-mode: "form" }
# This is your hands-on practice!

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

print("="*60)
print(" BUSINESS CASE: Predicting Sales from Advertising Spend")
print("="*60)

# Step 1: Create the data
ad_spend = np.array([1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000]).reshape(-1, 1)  # dollars
revenue = np.array([15000, 22000, 28000, 35000, 42000, 48000, 55000, 61000])  # dollars

print("\nüìä Historical Data:")
print("-"*40)
print("Ad Spend ($) | Revenue ($)")
print("-"*40)
for spend, rev in zip(ad_spend, revenue):
    print(f"  ${spend[0]:,}     | ${rev:,}")
print("-"*40)

# Step 2: Visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(ad_spend, revenue, color='green', s=150, alpha=0.6,
            edgecolors='darkgreen', linewidths=2)
plt.xlabel('Advertising Spend ($)', fontsize=13, fontweight='bold')
plt.ylabel('Revenue ($)', fontsize=13, fontweight='bold')
plt.title('Advertising Spend vs Revenue', fontsize=15, fontweight='bold', pad=15)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

# Step 3: Create and train the model
model = LinearRegression()
model.fit(ad_spend, revenue)

print(f"\nüìê Model Equation:")
print(f"   Revenue = ${model.intercept_:,.0f} + ${model.coef_[0]:.2f} √ó Ad Spend")
print(f"\nüí° What this tells us:")
print(f"   ‚Ä¢ Base revenue (no ads): ${model.intercept_:,.0f}")
print(f"   ‚Ä¢ Each $1 in ads generates: ${model.coef_[0]:.2f} in revenue")

# Step 4: Make predictions
new_ad_spend = np.array([[4500]])
predicted_revenue = model.predict(new_ad_spend)

print(f"\nüîÆ Prediction:")
print(f"   If you spend ${new_ad_spend[0][0]:,} on ads...")
print(f"   Expected revenue: ${predicted_revenue[0]:,.0f}")

# What ad spend would you need to hit $50,000 in revenue?
target_revenue = 50000
required_ad_spend = (target_revenue - model.intercept_) / model.coef_[0]
print(f"\nüéØ Reverse Calculation:")
print(f"   To reach ${target_revenue:,} in revenue...")
print(f"   You need to spend: ${required_ad_spend:,.0f} on ads")

# Step 5: Evaluate the model
train_predictions = model.predict(ad_spend)
r2 = r2_score(revenue, train_predictions)
mae = mean_absolute_error(revenue, train_predictions)

print(f"\nüìà Model Performance:")
print(f"   R¬≤ Score: {r2:.3f} ({'Excellent!' if r2 > 0.95 else 'Good!' if r2 > 0.8 else 'Okay'})")
print(f"   Mean Absolute Error: ${mae:,.0f}")
print(f"   Average prediction is off by ${mae:,.0f}")

# Step 6: Visualize predictions
plt.figure(figsize=(12, 6))
plt.scatter(ad_spend, revenue, color='green', s=150, alpha=0.7,
           edgecolors='darkgreen', linewidths=2, label='Actual Historical Data')

# Prediction line
ad_range = np.linspace(0, 10000, 100).reshape(-1, 1)
predicted_revenue_range = model.predict(ad_range)
plt.plot(ad_range, predicted_revenue_range, color='red', linewidth=3,
         label='Model Predictions')

# Highlight your specific prediction
plt.scatter(new_ad_spend, predicted_revenue, color='gold', s=400,
           marker='*', edgecolors='orange', linewidths=2,
           label=f'Prediction: ${predicted_revenue[0]:,.0f}', zorder=5)

plt.xlabel('Advertising Spend ($)', fontsize=13, fontweight='bold')
plt.ylabel('Revenue ($)', fontsize=13, fontweight='bold')
plt.title('Advertising ROI Prediction Model', fontsize=15, fontweight='bold', pad=15)
plt.legend(fontsize=11, loc='lower right')
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

# Calculate ROI
roi = ((predicted_revenue[0] - new_ad_spend[0][0]) / new_ad_spend[0][0]) * 100
revenue_per_dollar = predicted_revenue[0] / new_ad_spend[0][0]

print(f"\n$$ Return on Investment (ROI) Analysis:")
print(f"   ROI: {roi:.1f}%")
print(f"   For every $1 spent on ads:")
print(f"   ‚Üí You get ${revenue_per_dollar:.2f} in revenue")
print(f"   ‚Üí Net profit per dollar: ${revenue_per_dollar - 1:.2f}")

print("\n" + "="*60)
print(" EXERCISE COMPLETE! ‚úÖ")
print("="*60)
```
